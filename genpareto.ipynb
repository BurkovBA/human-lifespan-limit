```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import genpareto


# Generate data points for the x-axis (range of values for the density plot)
x = np.linspace(-1, 20, 500)

# Parameters for the Generalized Pareto Distribution
shapes = [-0.25, 0, 4]
colors = ['blue', 'green', 'red']
labels = ['Shape $\gamma$ = -0.25', 'Shape $\gamma$ = 0', 'Shape $\gamma$ = 4']

# Plot the densities for each shape parameter
plt.figure(figsize=(20, 12))
for shape, color, label in zip(shapes, colors, labels):
    # Calculate the density values for the current shape parameter
    density = genpareto.pdf(x, shape)

    # Plot the density curve
    plt.plot(x, density, color=color, label=label)

# Add labels and title
plt.xlabel("X-axis", fontsize=20)
plt.ylabel("Density", fontsize=20)
plt.title("Generalized Pareto Distribution with Various Shape Parameters", fontsize=20)

# Add legend
plt.legend(fontsize=18)

# Show the plot
plt.show()
```
